```python
from sklearn.preprocessing import normalize
from sklearn.model_selection import train_test_split

import numpy as np

def is_nlos(y):
    if y in [1,3,4]:
        return 1
    else:
        return 0

X_data = [eval(x) for x in df.cir.values]
y_data = np.asarray([is_nlos(y) for y in df.to_id])

# X_data = normalize(np.real(np.fft.fft(X_data)))
X_data = normalize(X_data)

X_train, X_test, y_train, y_test = train_test_split(X_data, y_data, test_size=0.2, random_state=0)
```

```python


model = keras.models.Sequential()
# model.add(layers.Conv1D(8,32,activation='relu', input_shape=(1016,1)))
# model.add(layers.Dropout(0.2))
# model.add(layers.MaxPool1D(16))
# model.add(layers.Conv1D(8,32,activation='relu', input_shape=(1016,1)))
# model.add(layers.Dropout(0.2))
# model.add(layers.MaxPool1D(16))
# model.add(layers.Flatten())
# model.add(layers.Dense(16, activation='relu'))
# model.add(layers.Dropout(0.2))
# model.add(layers.Dense(16, activation='relu'))
# model.add(layers.Dropout(0.2))
model.add(layers.Dense(32, activation='relu'))
model.add(layers.Dropout(0.2))
model.add(layers.Dense(32, activation='relu'))
model.add(layers.Dropout(0.2))
model.add(layers.Dense(32, activation='relu'))
model.add(layers.Dropout(0.2))
model.add(layers.Dense(1))
```

```python
model.summary()
```

```python
model.compile(optimizer=optimizers.Adam(1e-4), metrics=['acc'], loss="binary_crossentropy")
history = model.fit(X_train, y_train, epochs=200, validation_data=(X_test, y_test), batch_size=16)
```

```python
import matplotlib.pyplot as plt

plt.plot(history.history['acc'])
plt.plot(history.history['val_acc'])
```

```python
import matplotlib.pyplot as plt

plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
```

```python
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier, ExtraTreesClassifier

clf = SVC()
clf.fit(X_train, y_train)
clf.score(X_test, y_test)
```
